```ruby
# Balance a tree as BST

def create_balanced_bst(arr)
    if arr.length == 0
        nil
    elsif arr.length == 1
        TreeNode.new(arr.last)
    else
        idx = arr.length/2
        TreeNode.new(arr[idx], create_balanced_bst(arr[0..idx-1]), create_balanced_bst(arr[idx+1..arr.length-1]))
    end
end

def inorder_traversal(root, res = [])
    return if root.nil?
    inorder_traversal(root.left, res)
    res.push(root.val)
    inorder_traversal(root.right, res)
end

def balance_bst(root)
    res = []
    inorder_traversal(root, res)
    create_balanced_bst(res)
end
```

```ruby
# Binary Tree Level Order Traversal

#ITERATIVE

def level_order(root)
    
    if root.nil?
      puts "Queue is empty"
      return
    end

    queue = [root]
    # queue.enqueue(root)

    while !queue.empty? do
      node = queue.shift
      puts node.val

      unless node.left.nil?
        queue.push(node.left)
      end

      unless node.right.nil?
        queue.push(node.right)
      end
    end
end

def level_order(root)
  result = []
  return result if root.nil?

  queue = []
  queue << root

  until queue.empty?
    level_size = queue.length
    level = []
    level_size.times do
      node = queue.shift
      level << node.val
      queue << node.left unless node.left.nil?
      queue << node.right unless node.right.nil?
    end
    result << level
  end

  result
end

# RECURSIVE

def level_order(root, result = [], level = 0)
  return result unless root

  result << [] if result.length == level
  result[level] << root.val
  level_order(root.left, result, level + 1)
  level_order(root.right, result, level + 1)
end
```

```ruby
# Binary Tree implementation

class TreeNode
    attr_accessor :vlue, :left, :right
    #node is initialized with a value, and two children are nil
    def initialize(value)
        @value = value
        @left = nil
        @right = nil
    end
end

class BST
    attr_accessor :root, :size
    def initialize()
        @root = nil
        @size = 0;
    end
end

def insert(value)
    if @root == nil
        @root = TreeNode.new(value)
    else
        curr_node = @root
        previous_node = @root
        #while loop helps finding the position of insertion
        while curr_node != nil
            previous_node = curr_node
            if value < curr_node.value
                curr_node = curr_node.left
            else
                curr_node = curr_node.right
            end
        end
        if value < previous_node.value
            previous_node.left = TreeNode.new(value)
        else
            previous_node.right = TreeNode.new(value)
        end
    end
    @size += 1
end

def find_max(node = self.root)
    if node == nil
        return nil
    elsif node.right == nil
        return node
    end
    return find_max(node.right)
end

def find_min(node = self.root)
    if node == nil
        return nil
    elsif node.left == nil
        return node
    end
    return find_min(node.left)
end

def contains?(value, node = self.root)
    if node == nil
        return false
    elsif value < node.value
        return contains?(value, node.left)
    elsif value > node.value
        return contains?(value, node.right)
    else
        return true
    end
end

```
